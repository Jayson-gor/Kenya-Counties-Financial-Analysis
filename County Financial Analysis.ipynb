## Pending Bills for the Counties as of June 30, 2022

```python
import tabula
import pandas as pd

# Path to the PDF file
pdf_path = "C:/Users/gorja/Downloads/Counties-Sep-2022-web.pdf"

# Extract tables from the specified pages
tables_page_33 = tabula.read_pdf(pdf_path, pages=33, multiple_tables=True)
tables_page_34 = tabula.read_pdf(pdf_path, pages=34, multiple_tables=True)

# Check if tables were extracted and handle empty cases
if tables_page_33 and tables_page_34:
    # Assuming the relevant tables are the first ones on each page
    table_33 = tables_page_33[0]
    table_34 = tables_page_34[0]
    
    # Combine the extracted tables into a single DataFrame
    combined_df = pd.concat([table_33, table_34], ignore_index=True)
    
    # Select only the first and last columns
    filtered_df = combined_df.iloc[:, [0, -1]]
    
    # Display the filtered DataFrame
    pd.set_option('display.max_rows', None)  # Show all rows
    pd.set_option('display.max_columns', None)  # Show all columns
    print(filtered_df)
    
    # Save the filtered DataFrame to a CSV file
    output_path = 'Pending_Bills_for_the_Counties_as_of_June_2021_2022_filtered.csv'
    filtered_df.to_csv(output_path, index=False)
    print(f"Filtered DataFrame saved to '{output_path}'.")
else:
    print("No tables found on page 33 or 34.")
```

## DATA CLEANING

```python
import pandas as pd

# Load the data
input_path = "C:/Users/gorja/OneDrive/Documents/Desktop/Counties Project/County_Budget_Allocation_Expenditure_and_absorption Rate_2021_2022.csv"
df = pd.read_csv(input_path)

# Display the first few rows of the DataFrame
print("Initial DataFrame:")
print(df.head(50))

# Display data types
print("\nData types:")
print(df.dtypes)
```

```python
#Handling mising values
# Fill missing values with 0
df.fillna(0, inplace=True)
```

```python
# check missing values:
# Check for missing values using isnull()
missing_values = df.isnull()

# Display the resulting DataFrame
print(missing_values)
```

```python
# Assuming you have a DataFrame named 'df'
print("Data Types of Columns in DataFrame:")
print(df.dtypes)
```

```python
# Assuming you have a DataFrame named 'df'
df.rename(columns={'A': 'Budget Estimates (Kshs. Million) Rec', 'B': 'Budget Estimates (Kshs. Million) Dev', 'C=A+B': 'Budget Estimates (Kshs. Million) Total', 'D': 'Expenditure (Kshs. Million) Rec', 'E': 'Expenditure (Kshs. Million) Dev', 'Recurrent \nAbsorption \nRate (%)': 'Recurrent Absorption Rate (%)', 'Development \nAbsorption \nRate (%)':'Development Absorption Rate (%)', 'Overall Ab\nsorption Rate\n (%)':'Overall Absorption Rate (%)' }, inplace=True)

# Display the updated DataFrame
print(df.head())
```

```python
# Convert specified columns to numeric, handling any formatting issues
columns_to_convert = [
    'Budget Estimates (Kshs. Billion) Rec', 
    'Budget Estimates (Kshs. Billion) Dev', 
    'Budget Estimates (Kshs. Billion) Total',
    'Expenditure (Kshs. Billion) Rec', 
    'Expenditure (Kshs. Billion) Dev', 
    'Expenditure (Kshs. Billion) Total',
    'Recurrent Absorption Rate (%)', 
    'Development Absorption Rate (%)', 
    'Overall Absorption Rate (%)'
]

for column in columns_to_convert:
    if column in df.columns:
        df[column] = df[column].astype(str)  # Convert to string to handle .str methods
        df[column] = df[column].str.replace(',', '').str.strip()  # Remove commas and spaces
        df[column] = pd.to_numeric(df[column], errors='coerce')  # Convert to numeric


# Assuming you have a DataFrame named 'df'
print("Data Types of Columns in DataFrame:")
print(df.dtypes)
```

```python
### Save the file
output_path = 'Cleaned_County_Budget_Allocation_Expenditure_and_Absorption_Rate_2021_2022.csv'
df.to_csv(output_path, index=False)
print(f"Cleaned data saved to '{output_path}'.")
```
